```python
url_jan="201901-citibike-tripdata.csv/201901-citibike-tripdata.csv"
jan_df=pd.read_csv(url_jan)
jan_df.head()
```

```python
# Splitting the column 'title' into two columns 'Title' and 'Release Year'
split1= jan_df["starttime"].str.rsplit(" ", n = 1, expand = True)
jan_df["startday"] = split1[0]
jan_df["starttime"] = split1[1]

split2= jan_df["starttime"].str.rsplit(".", n = 1, expand = True)
jan_df["starttime"] = split2[0]
jan_df.head()
```
